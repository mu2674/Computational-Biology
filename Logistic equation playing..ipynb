```python
import matplotlib.pyplot as plt
import random
#Imports
```

```python
# set a maximum population cap. 
maxpop = 1000.0

#Create arrays to hold scatterplot points. 
xaxis = []
yaxis = []

# iterate over all possible cases of r (from 1.0 to 4.0, intervals of 0.001)
# Range does not work with decimals so went with thousands. 
for r in range(1000, 4000):

    r = r / 1000.0  # turn the r integer into a float, ex: 3006 = 3.006

    # generate a random starting population
    pop = random.randrange(2, 20)
    ##pop = 100    
    # this does weird stuff

    # find this population ratio xn
    xn = pop / maxpop


    # print("r: " + str(r))
    # print("pop: " + str(pop))
    # print("xn: " + str(xn))
    #Debugging above ^^^^
    
    # change how many times you want to test the same value of r
    for i in range(0, 1):

        # iterate over however many years
        for year in range(0, 200):

            # logistic map function
            xn1 = r * xn * (1 - xn)

            popnew = int(maxpop * xn1)

            # print("    popnew: " + str(year) + ": " + str(popnew))
            # print("    xn1: " + str(xn1))

            xn = xn1

        xaxis.append(r) # x coordinate
        yaxis.append(popnew) # y coordinate
#Play with size for different iterations. 
plt.figure(figsize=(10,10))
#Creates the plot but does not display. Displays internally.
plt.scatter(xaxis, yaxis)
#Displays plot. 
plt.show()
```
