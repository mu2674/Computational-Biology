```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 20000
e = 0.6 # lettuce fertility rate
f = 0.8 # chance that the rabbit and the lettuce meet
a = 0.6 # Bunny fertility rate
b = 1.2 # chance of bunny and fox meet
c = 0.5 # 
d = 1

dt = 0.001

possible_ds = [0.5, 1, 1.5, 2]

z = np.zeros(nsteps)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16, 5));
fig2 = plt.figure(figsize=(5, 5));
ax2 = plt.axes()

for itrial in range(len(possible_ds)):
   d = possible_ds[itrial]
   z[0] = 1
   x[0] = 1
   y[0] = 1

   for t in range(1, nsteps):
      z[t] = z[t - 1] + dt * (e * z[t - 1] - f * z[t - 1] * x[t - 1])
      x[t] = x[t - 1] + dt * (a * x[t - 1] * z[t - 1] - b * x[t - 1] * y[t - 1])
      y[t] = y[t - 1] + dt * (c * x[t - 1] * y[t - 1] - d * y[t - 1])
      # x[t] = x[t - 1] + dt * (a * x[t - 1] - b * x[t - 1] * y[t - 1])
      # y[t] = y[t - 1] + dt * (c * x[t - 1] * y[t - 1] - d * y[t - 1])

   ax1 = fig1.add_subplot(1, len(possible_ds), itrial + 1)
   ax1.plot(z, label='Lettuce');
   ax1.plot(x, label='Bunnies');
   ax1.plot(y, label='Foxes');
   ax1.legend(loc='upper right');

   ax2.plot(x, y, label=d);
   plt.xlabel('Bunnies');
   plt.ylabel('Foxes');

ax2.legend(loc='upper right');
plt.show()
```
